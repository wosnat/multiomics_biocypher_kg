```python
merge_df = pd.merge(ncbi_df, luca_df, on=['start', 'end', 'strand'], how='outer', suffixes=['_ncbi', '_luca'])
```

```python
merge_df.ROS_degradation.unique()
```

```python

cols = [ 'locus_tag_ncbi','locus_tag_luca', 'gene_ncbi', 'gene_luca',
       'product_ncbi', 'product_luca', #'is_signal_ncbi',  'is_signal_luca', 
        'Protease_Aminase',  
           'category','EC_number', 'EC_number_from_kegg', 'is_signal_ncbi', 'is_signal_luca',
       ]

with pd.ExcelWriter('protease_comparison_ncbi_luca.xlsx') as writer:  
    merge_df.loc[
        merge_df.category.isin(['Proteases', 'D-/L-Amino acid oxidases']) & 
        merge_df.Protease_Aminase.isin(['Protease/Aminase']), 
    cols].to_excel(writer, sheet_name='Common', index=False)

    merge_df.loc[
        ~merge_df.category.isin(['Proteases', 'D-/L-Amino acid oxidases']) & 
        merge_df.Protease_Aminase.isin(['Protease/Aminase']), 
    cols].to_excel(writer, sheet_name='NCBI only', index=False)

    merge_df.loc[
        merge_df.category.isin(['Proteases', 'D-/L-Amino acid oxidases']) & 
        ~merge_df.Protease_Aminase.isin(['Protease/Aminase']), 
    cols].to_excel(writer, sheet_name='LUCA only', index=False)
```

```python
for (index, locus_tag, start, end, strand) in missing_ncbi.itertuples():
    delta = 500
    print(locus_tag, start, end, strand)
    print(ncbi_df.loc[
              ncbi_df.start.between(start-delta, start+delta) &
              ncbi_df.end.between(end-delta, end+delta) &
              (ncbi_df.strand == strand),
              ['locus_tag', 'start', 'end', 'strand', 'product', 'Protease_Aminase']

          ])
    print()
```
