```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pprint
sns.set(style="white", context='paper')
```

```python
import os
dpath = 'signalP output'
for i in os.listdir(dpath):
    fpath = os.path.join(dpath, i, 'prediction_results.txt')
    print(fpath)
```

```python
def _read_(i):
    fpath = os.path.join(dpath, i, 'prediction_results.txt')
    df = pd.read_csv(fpath, sep='\t', skiprows=1)
    return df

df = pd.concat([_read_(i) for i in os.listdir(dpath)], ignore_index=True)
df
    
```

```python
df[['protein_id', 'product']] = df['# ID'].str.split(' ', n=1, expand=True)
df
```

```python
df[['protein_id', 'product',  'Prediction']].to_csv('data/HOT1A3_signalP.csv', index=False)
```
